```python
def MakeSequence(oligo):
    control_plasmid = plasmid1 + oligo+plasmid2
    control_seq_right = control_plasmid+right_seq
    control_seq_right = control_seq_right[:50000]
    control_seq = left_seq+control_seq_right
    return control_seq
```

```python
#Make signal predcition with Puffin-D
prediciton_puffin_d = np.zeros((len(data), 10, 10000))

for i in range(len(data)):
    print(i)
    oligo = data['Oligo_sequence'].values[i]
    seq = MakeSequence(oligo)
    pred = puffin_d.predict(seq)
    pred = pred[0, :, 50000-5000:50000+5000]
    
    prediciton_puffin_d[i, :, :] = pred
    
```

```python
targeti = 4
efp_means = np.zeros((len(data)))
data_Exp = (10**prediciton_puffin_d)-1
for i in range(len(data)):
    pred = data_Exp[i,targeti,:]
    
    efp_lim = (len(plasmid1)+len(oligo)+len(plasmid2[:100]))
    efp_means[i] = np.mean(pred[5000:5000+efp_lim])
```

```python
mcherry = []
for i in range(0, len(data)):
    
    mcherry.append(np.mean(data_Exp[i, (-1*targeti)-1, 4427-200:4427]))
```

```python
green_pred = efp_means.copy()
plt.scatter(range(15753), np.array(green_pred)[np.argsort(np.array(green_pred))], s=1, c='green')
plt.scatter(range(15753), np.array(mcherry)[np.argsort(np.array(green_pred))], s=1, c='red')
```

```python
data_filtered = data['Mean_log10'].values[~np.isnan(data['Mean_log10'].values)]
filtr = ~np.isnan(data['Mean_log10'].values)
pred_filtered = green_pred[filtr]/np.array(mcherry)[filtr]


print(scipy.stats.spearmanr(data_filtered, np.log(pred_filtered)))
```

```python
def MakeSequenceGenome(oligo):
    
    control_seq = left_seq[-825:]+(oligo+right_seq)[:825]
    return control_seq
```

```python
#Make signal predcition with Puffin

prediciton_all = np.zeros((len(data), 10,1000))

for i in range(len(data)):
    print(i)
    oligo = data['Oligo_sequence'].values[i]
    
    seq = MakeSequenceGenome(oligo)
    prediction = puffin.predict(seq)

    prediciton_all[i, :,:] = prediction.values[2:,:]
```

```python
pred = np.exp(prediciton_all)-1
pred = np.mean(pred[:,0,500:900], axis=-1)


filtr = ~np.isnan(data['Mean_log10'].values)

pred2 = pred[filtr]
exper = data['Mean_log10'].values[filtr]

scipy.stats.spearmanr(exper, np.log(pred2))
```
